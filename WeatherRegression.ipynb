```python
import pandas as pd
df = pd.read_csv("/content/weatherHistory.csv")
df[:5]
```

```python
def map_categorical_to_numerical(categorical_data):
    unique_categories = set(categorical_data)
    mapping = {category: i for i, category in enumerate(unique_categories)}
    return mapping
```

```python
def string_to_number(string):
  return round(float(string),2)
```

```python
summary_numerical = map_categorical_to_numerical(df["Summary"])
precip_type_numerical = map_categorical_to_numerical(df["Precip Type"])
```

```python
import numpy as np

new_summary = np.array([])
for summary in df["Summary"]:
  new_summary = np.append(new_summary,summary_numerical[summary])

new_summary[:10]
```

```python
df = df.drop('Summary', axis=1)
df["Summary"] = new_summary
df[:5]
```

```python
new_precip = np.array([])
for precip in df["Precip Type"]:
  new_precip = np.append(new_precip,precip_type_numerical[precip])

new_precip[:10]
```

```python
df = df.drop('Precip Type', axis=1)
df = df.drop('Daily Summary', axis=1) 
df["Precip Type"] = new_precip
df[:5]
```

```python
from datetime import datetime
date_format = "%Y-%m-%d %H:%M:%S.%f %z"
new_times = np.array([])

for time in df["Formatted Date"]:
  new_times = np.append(new_times,datetime.strptime(time, date_format))

new_times[:5]
```

```python
df = df.drop('Formatted Date', axis=1)
df["Formatted Date"] = new_times
df = df.dropna().drop("Formatted Date", axis=1).astype(np.float32)
df[:5]
```

```python
full_labels = df["Apparent Temperature (C)"]
full_vals = df.drop("Apparent Temperature (C)", axis=1)
full_labels[:3], full_vals[:3]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(full_vals, full_labels, 
                                                    test_size=0.2, random_state=42)
X_train[:3],X_test[:3]
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics = ["mae"])

model.fit(X_train,y_train,
          epochs=20,
          verbose=1)
```

```python
model.evaluate(X_test,y_test)
```
